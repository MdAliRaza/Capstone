```python
import pandas as pd
csv_data = pd.read_csv(r'C:\Users\ASUS\Desktop\python\capstone\olympix_data.csv')
csv_data.head()
```

```python

import mysql.connector as mysql
from mysql.connector import Error
try:
    conn = mysql.connect(host='localhost', 
                        user='root',        # give your username
                        password='password')# # give your password
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("CREATE DATABASE olympics")
        print("Database is created")
except Error as e:
    print("Error while connecting to MySQL", e)
```

```python
try:
    conn = mysql.connect(host='localhost', 
                         database='olympics', 
                         user='root', 
                         password='password')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS olympix;')
        print('Creating table....')
# in the below line please pass the create table statement which you want #to create
        cursor.execute("CREATE TABLE olympix(name VARCHAR(100) NOT NULL,age integer(10) NOT NULL,country VARCHAR(100) NOT NULL,year varchar(10) NOT NULL,date_given DATE NOT NULL,sports VARCHAR(100) NOT NULL,gold_medal integer(5) NOT NULL,silver_medal integer(5) NOT NULL,brone_medal integer(5) NOT NULL,total_medal integer(5) NOT NULL)")
        print("Table is created....")
        #loop through the data frame
        for i,row in csv_data.iterrows():
            #here %S means string values 
            sql = "INSERT INTO olympics.olympix VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)

```
